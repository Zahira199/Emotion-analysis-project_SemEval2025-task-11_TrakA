```python
import pandas as pd
import re

# Charger le fichier CSV
df = pd.read_csv("/content/ary (2).csv")
df.head()
```

```python
# Fonction pour nettoyer le texte
import re

def clean_text(text):
    # Enlever les mentions, hashtags et URL
    text = re.sub(r'@<username>', '', text)
    text = re.sub(r'##', '', text)
    text = re.sub(r'URL', '', text)

    # Emojis traduits vers des émotions spécifiques
    emoji_mapping = {
        "😂": "فرح",       # joy
        "😘": "فرح",       # joy
        "😍": "فرح",       # joy
        "✨": "مفاجأة",     # surprise
        "🔥": "مفاجأة",     # surprise
        "😒": "حزن",       # sadness
        "☹️": "حزن",       # sadness
        "😭": "حزن",       # sadness
        "🤔": "حزن",       # interpreted as sadness/reflection
        "😢": "حزن",       # sadness
        "😤": "غضب",       # anger
        "💩": "اشمئزاز"     # disgust
    }

    for emoji, arabic_emotion in emoji_mapping.items():
        text = text.replace(emoji, arabic_emotion)

    return text


# Appliquer la fonction de nettoyage
df["cleaned_text"] = df["text"].apply(clean_text)
df.head()
```

```python
def arabizi_to_arabic(text):
    mapping = [
        ("3'", "غ"), ("3", "ع"), ("7'", "خ"), ("7", "ح"), ("9", "ق"), ("6", "ط"),
        ("2", "أ"), ("5", "خ"), ("gh", "غ"), ("kh", "خ"), ("ch", "ش"), ("sh", "ش"),
        ("d", "د"), ("h", "ه"), ("b", "ب"), ("f", "ف"), ("j", "ج"), ("k", "ك"),
        ("l", "ل"), ("m", "م"), ("n", "ن"), ("r", "ر"), ("s", "س"), ("t", "ت"),
        ("z", "ز"), ("w", "و"), ("y", "ي"), ("a", "ا"), ("e", "ي"), ("o", "و")
    ]

    text = text.lower()
    for latin, arabic in mapping:
        text = text.replace(latin, arabic)
    return text
```

```python

# Appliquer la translittération uniquement aux lignes contenant de l'arabizi
df['text_transliterated'] = df['cleaned_text'].apply(arabizi_to_arabic)

# Afficher un échantillon de résultats
df[['cleaned_text', 'text_transliterated']].head(10)
```

```python
import re

def arabizi_to_arabic(text):
    mapping = [
        ("3'", "غ"), ("7'", "خ"), ("gh", "غ"), ("kh", "خ"), ("sh", "ش"), ("ch", "ش"),
        ("3", "ع"), ("7", "ح"), ("9", "ق"), ("6", "ط"), ("2", "أ"), ("5", "خ"),
        ("d", "د"), ("h", "ه"), ("b", "ب"), ("f", "ف"), ("j", "ج"), ("k", "ك"),
        ("l", "ل"), ("m", "م"), ("n", "ن"), ("r", "ر"), ("s", "س"), ("t", "ت"),
        ("z", "ز"), ("w", "و"), ("y", "ي"), ("a", "ا"), ("e", "ي"), ("o", "و")
    ]
    # Trier pour d'abord appliquer les remplacements longs
    mapping = sorted(mapping, key=lambda x: -len(x[0]))

    def convert_word(word):
        if re.fullmatch(r"[a-z0-9']+", word.lower()):
            for latin, arabic in mapping:
                word = word.replace(latin, arabic)
        return word

    words = text.split()
    converted = [convert_word(w) for w in words]
    return " ".join(converted)
```

```python

# Appliquer la translittération uniquement aux lignes contenant de l'arabizi
df['text_transliterated'] = df['cleaned_text'].apply(arabizi_to_arabic)

# Afficher un échantillon de résultats
df[['cleaned_text', 'text_transliterated']].head(10)
```

```python
def arabizi_to_arabic(text):
    mapping = [
        ("3'", "غ"), ("3", "ع"), ("7'", "خ"), ("7", "ح"), ("9", "ق"), ("6", "ط"),
        ("2", "أ"), ("5", "خ"), ("gh", "غ"), ("kh", "خ"), ("ch", "ش"), ("sh", "ش"),
        ("aa", "ا"), ("oo", "و"), ("ou", "و"), ("i", "ي"), ("u", "و"), ("a", "ا"),
        ("e", "ي"), ("o", "و"), ("é", "ي"), ("è", "ي"), ("ê", "ي"), ("â", "ا"),
        ("û", "و"), ("î", "ي"), ("'", ""), ("b", "ب"), ("f", "ف"), ("j", "ج"),
        ("k", "ك"), ("l", "ل"), ("m", "م"), ("n", "ن"), ("r", "ر"), ("s", "س"),
        ("t", "ت"), ("z", "ز"), ("w", "و"), ("y", "ي"),
    ]

    text = text.lower()
    for latin, arabic in mapping:
        text = text.replace(latin, arabic)

    text = re.sub(r'(.)\1+', r'\1\1', text)  # normalise les doublons
    return text
```

```python
import re

def keep_only_arabic(text):
    # Garder uniquement les lettres arabes et espaces
    text = re.sub(r'[^\u0600-\u06FF\u0750-\u077F\u08A0-\u08FF\u0660-\u0669\s]', '', text)
    return text
```

```python
# Appliquer la translittération uniquement aux lignes contenant de l'arabizi
df['cleaned_text'] =df['text_transliterated']
df['text_transliterated'] = df['cleaned_text'].apply(keep_only_arabic)

# Afficher un échantillon de résultats
df[['cleaned_text', 'text_transliterated']].head(10)
df[['text_transliterated']].to_csv("keep_only_arabic.csv", index=False)
```
